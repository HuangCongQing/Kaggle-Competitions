# Python数据分析建模--分类与预测

## 项目简介
* [项目地址](https://www.kaggle.com/uciml/glass)
* 任务：根据玻璃材料的9个属性对未知的玻璃类别进行分类和预测

# 1. 准备阶段
## 加载相关模块和库

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.neighbors import KNeighborsClassifier
import numpy as np

%matplotlib inline
```

## 加载并查看数据

```python
dataset_path = './dataset/glass.csv'
data = pd.read_csv(dataset_path)
feature_names = data.columns[:-1].tolist()
print('数据形状：{}'.format(data.shape))
print('共有{}条记录'.format(data.shape[0]))
print('共有{}个特征：{}'.format(len(feature_names), feature_names))

print()
print('数据预览：')
data.head()

print('属性类型：')
data.dtypes
```

# 2. 数据分析及可视化
## 查看统计信息

```python
print('数据统计信息：')
data.describe()
```

通过简单的查看，发现特征数据的范围不一致，对模型的学习过程有一定影响。比如RI和Si的均值相差过大，不在同一个数值范围中。

接下来查看各类别的数据记录个数。

```python
data['Type'].value_counts()
```

类别分布不平衡，其中类别1和2占了大多数。

## 数据可视化
查看特征的数据分布情况

```python
for feature in feature_names:
    skew = data[feature].skew()
    sns.distplot(data[feature], label='skew = %.3f' %(skew))
    plt.legend(loc='best')
    plt.show()
```

```python
sns.boxplot(data[feature_names])
plt.show()
```

多变量数据可视化

```python
plt.figure(figsize=(8, 8))
sns.pairplot(data[feature_names])
plt.show()
```

# 3. 数据处理
## 分割数据集

```python
X = data[feature_names].values
y = data['Type'].values

seed = 10
test_size = 0.2
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = test_size , random_state = seed)
```

# 4. 数据建模

```python
# 选择模型，交叉验证
k_range = range(1, 31)
cv_scores = []
print('交叉验证：')
for k in k_range:
    knn = KNeighborsClassifier(k)
    scores = cross_val_score(knn, X_train, y_train, cv=10, scoring='accuracy')
    score_mean = scores.mean()
    cv_scores.append(score_mean)
    print('%i: %.4f' % (k, score_mean))

best_k = np.argmax(cv_scores) + 1
print('最优K: {}'.format(best_k))

plt.plot(k_range, cv_scores)
plt.xlabel('K')
plt.ylabel('Accuracy')
plt.show()
```

# 5. 分类和预测

```python
# 训练模型
knn_model = KNeighborsClassifier(best_k)
knn_model.fit(X_train, y_train)
print('测试模型，准确率：{}'.format(knn_model.score(X_test, y_test)))
```
